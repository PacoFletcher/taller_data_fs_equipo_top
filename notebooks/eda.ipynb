```python
import re


def limpiar_features(texto):
    if pd.isna(texto):
        return texto
    # Eliminar: "X habs.", "X habitaciones", "X baño", "X baños", "XXX m²"
    return re.sub(r'\b\d+\s*habs?\.?,?|\b\d+\s*habitaciones?,?|\b\d+\s*bañ[oa]s?\.?,?|\b\d+\s*m²,?', '', texto).strip(', ').strip()

# Aplicar al DataFrame
df3['features_limpio'] = df3['features'].apply(limpiar_features)
```

```python
df3
```

```python
def filtrar_features_sin_numeros(texto):
    if pd.isna(texto):
        return ''
    # Separar por coma y quitar espacios
    items = [item.strip() for item in texto.split(',')]
    # Eliminar los que contienen números
    filtrados = [item for item in items if not re.search(r'\b\d+[ºª]?', item)]
    return ','.join(filtrados)

# Aplicar al DataFrame
df3['features_limpio'] = df3['features'].apply(filtrar_features_sin_numeros)
df3
```

```python
df3['features_list'] = df3['features_limpio'].apply(lambda x: [item.strip() for item in x.split(',') if item.strip()])
features_dummies = df3['features_list'].explode().str.get_dummies().groupby(level=0).sum()
df3 = pd.concat([df3, features_dummies], axis=1)
df3
```

```python
# Paso 1: Calcular el precio medio por location_name, excluyendo los -1
precio_medio_por_localidad = (
    df3[df3["price"] != -1]
    .groupby("location_name")["price"]
    .mean()
    .round(2)
)

# Paso 2: Crear columna con el precio medio correspondiente a cada fila
df3["location_mean_price"] = df3["location_name"].map(precio_medio_por_localidad)

# Paso 3: Reemplazar -1 en 'price' por el precio medio correspondiente
df3.loc[df3["price"] == -1, "price"] = df3.loc[df3["price"] == -1, "location_mean_price"]
```
